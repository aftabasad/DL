```python
import numpy as np
import matplotlib.pyplot as plt
```

### Softmax function implementation

```python
def softmax(a):
    return np.exp(a)/sum(np.exp(a))
```

```python
a = [-1.8, -2.0, -1.0, 4.5, 3.1, 3.0, 2.0]
y = softmax(a)
y
```

```python
np.max(y)
```

```python
np.argmax(y)
```

### sigmoid function and its derivative

```python
def sigmoid(x):
    s=1/(1+np.exp(-x))
    ds=s*(1-s)  
    return s,ds
```

```python
x=np.arange(-6,6,0.01)
sigmoid(x)[0].size,sigmoid(x)[1].size
```

```python
plt.plot(x,sigmoid(x)[0], color="#307EC7", linewidth=3, label="sigmoid")
plt.plot(x,sigmoid(x)[1], color="#9621E2", linewidth=3, label="derivative")
plt.legend(loc="upper right", frameon=False)
plt.show()
```

### tanh function and its derivative

```python
def tanh(x):
    t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    dt=1-t**2
    return t,dt
```

```python
z=np.arange(-4,4,0.01)
tanh(z)[0].size,tanh(z)[1].size
```

```python
plt.plot(z,tanh(z)[0], color="#307EC7", linewidth=3, label="sigmoid")
plt.plot(z,tanh(z)[1], color="#9621E2", linewidth=3, label="derivative")
plt.legend(loc="upper right", frameon=False)
plt.show()
```

### ReLU function and its derivative

```python
def relu(x):
    r = np.maximum(0,x)
    dr = (x > 0).astype(int)
    return r,dr
```

```python
r=np.arange(-4,4,0.01)
relu(r)[0].size,relu(r)[1].size
```

```python
plt.plot(r,relu(r)[0], color="#307EC7", linewidth=3, label="sigmoid")
plt.plot(r,relu(r)[1], color="#9621E2", linewidth=3, label="derivative")
plt.legend(loc="upper right", frameon=False)
plt.show()
```
